```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import tensorflow as tf
from keras.models import load_model

def load_and_preprocess_image(image_path, target_size=(64, 64)):
    image = Image.open(image_path)
    image = image.resize(target_size)
    image = np.array(image) / 255.0
    return image

# Directory containing background images
backgrounds_dir = '/content/drive/MyDrive/GAN/NotMineR2/'

background_files = os.listdir(backgrounds_dir)

# Path to the foreground image
foreground_path = '/content/drive/MyDrive/GAN/Foreground_Images/Copy of gan_plot_e1100.jpg' #Change

foreground_image = load_and_preprocess_image(foreground_path)

# Number of foreground images to use
num_foreground_images = 1

# Iteration number
i=500

# Iterate over each background image
for background_file in background_files:
    background_path = os.path.join(backgrounds_dir, background_file)
    background_image = load_and_preprocess_image(background_path)

    foreground_image_resized = tf.image.resize(foreground_image, (background_image.shape[0], background_image.shape[1]))

    blended_image = foreground_image_resized * 0.4 + background_image * 0.6  # Adjust blending ratio as needed

    plt.imshow(blended_image)
    path = '/content/drive/MyDrive/GAN/Final_Images/'
    filename = 'Mine2_e%03d.jpg' % (i)         #Change
    plt.savefig(path + filename)
    plt.axis('off')
    i+=1
```
